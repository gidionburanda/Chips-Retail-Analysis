```python
present_day = df21['DATE'].max() + 1
present_day
```

```python
recency_df= pd.DataFrame(df21.groupby(by='LYLTY_CARD_NBR', as_index=False)['DATE'].max())
recency_df
```

```python
recency_df['Recency']= recency_df['DATE'].apply(lambda x: (present_day - x))
recency_df
```

```python
frequency_df = pd.DataFrame(df21.groupby(["LYLTY_CARD_NBR"]).agg({"TXN_ID":"nunique"}).reset_index())
frequency_df.rename(columns={"TXN_ID":"Frequency"}, inplace=True)
```

```python
monetary_df = df21.groupby('LYLTY_CARD_NBR', as_index=False)['TOT_SALES'].sum()
monetary_df.columns = ['LYLTY_CARD_NBR', 'Monetary']
```

```python
RF_df = recency_df.merge(frequency_df, on='LYLTY_CARD_NBR')
RFM_df = RF_df.merge(monetary_df, on='LYLTY_CARD_NBR').drop(columns='DATE')
RFM_df
```

```python
segment_df = pd.merge(seg_df,RFM_df , on='LYLTY_CARD_NBR', how='inner')
segment_df
```

```python
sns.boxplot(data=segment_df, x='PREMIUM_CUSTOMER', y='Recency')
plt.title("Recency Distribution by Segment")
plt.show()
```

```python
import re
from collections import Counter
```

```python
s_df = segfu_df.loc[segfu_df['PREMIUM_CUSTOMER'] == 'Premium' ]
desired_premium_comments_df = s_df[['PREMIUM_CUSTOMER', 'PROD_NAME']].copy()
print(desired_premium_comments_df)
```

```python
def extract_brands(comment):
    pattern = r'\b(?:Cheetos|Lays|Aioli|RRD|Kettle|Doritos|Tostitos|Smiths|Smith|WW|Infuzions|Cheezels|Red Rock Deli|Infzns|Dorito|Twisties|Grain Waves|Burger Rings|NCC|GrnWves|Woolworths|Salsa)\b'

    brands = re.findall(pattern, comment, flags=re.IGNORECASE)

    return brands
all_brands = desired_premium_comments_df['PROD_NAME'].apply(lambda x: extract_brands(x)).tolist()
all_brands
```

```python
all_brands_flat = [brand for sublist in all_brands for brand in sublist]

brand_counts = Counter(all_brands_flat)
brand_counts
```
